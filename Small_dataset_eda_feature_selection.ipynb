```python
levCols = [#"year", 
           "manufacturer", "make", "condition", "cylinders", "fuel", #"odometer", 
            "title_status", "transmission", "drive", "car_size", "car_type", "paint_color"]#, "lat", "long"]

from sklearn import preprocessing
le = preprocessing.LabelEncoder()
dfcars2 = dfcars.copy()
for col in levCols:
    dfcars2[col] = le.fit_transform(dfcars[col])
```

```python
dfcars2.head()
```

```python
from scipy import stats
from sklearn import feature_selection

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

minfos=[]
pearsons = []
for var in levCols:
    print("\n", var)
    pearsons.append(stats.pearsonr(dfcars2["price"], dfcars2[var]))
    print("Pearson", stats.pearsonr(dfcars2["price"], dfcars2[var]))
    print("Mutual info", feature_selection.mutual_info_classif(dfcars2[var].values.reshape(-1,1),dfcars["price"].values))
    minfos.append(feature_selection.mutual_info_classif(dfcars2[var].values.reshape(-1,1),dfcars["price"].values))
    print("Chi2", feature_selection.chi2(dfcars2[var].values.reshape(-1,1),dfcars["price"].values))
    print("Anova", feature_selection.f_classif(dfcars2[var].values.reshape(-1,1),dfcars["price"].values))

imp = list(zip(minfos, levCols))
imp.sort(reverse=True)
imp
```
